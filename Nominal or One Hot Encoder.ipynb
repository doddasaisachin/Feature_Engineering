```python
# Nominal/One hot encoding
```

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df
```

```python
enc=OneHotEncoder()
```

```python
coded=enc.fit_transform(df[['color']]).toarray()
coded
```

```python
coded_df=pd.DataFrame(coded,columns=enc.get_feature_names_out())
coded_df
```

```python
enc.transform([['red']]).toarray()
```

```python
pd.concat([df,coded_df],axis=1)
```

```python
sex_model=OneHotEncoder()
```

```python
coded_sex=sex_model.fit_transform(df[['sex']]).toarray()
coded_sex
```

```python
sex_model.get_feature_names_out()
```

```python
coded_sex_df=pd.DataFrame(coded_sex,columns=sex_model.get_feature_names_out())
coded_sex_df.head()
```

```python
pd.concat([df,coded_sex_df],axis=1)
```

```python
smoke_model=OneHotEncoder()
smoke_array=smoke_model.fit_transform(df[['smoker']]).toarray()
smoke_array
```

```python
smoke_df=pd.DataFrame(smoke_array,columns=smoke_model.get_feature_names_out())
smoke_df.head()
```

```python
pd.concat([df,smoke_df],axis=1)
```

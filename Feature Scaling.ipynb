```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Normalization
```

```python
import numpy as np
```

```python
tot_bill_mean=np.mean(tips['total_bill'])
tot_bill_mean
```

```python
tot_bill_std=tips['total_bill'].std()
tot_bill_std
```

```python
normalized_data=[]
for i in list(tips['total_bill']):
    z_score=(i-tot_bill_mean)/(tot_bill_std)
    normalized_data.append(z_score)
normalized_data
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
sns.histplot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(tips[['total_bill']])
```

```python
scaler.transform(tips[['total_bill']])
```

```python
scaler.fit_transform(tips[['total_bill']])
```

```python
import pandas as pd
pd.DataFrame(scaler.fit_transform(tips[['total_bill','tip']]),columns=['scaled_bill','scaled_tip'])
```

```python
# Min-Max Scaler
```

```python
df=sns.load_dataset('taxis')
df.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit_transform(df[['distance','fare','total']])
```

```python
scaled_df=pd.DataFrame(scaler.fit_transform(df[['distance','fare','total','tip']]),columns=['scaled_dist','scaled_fare','scaled_tot','scaled_tip'])
```

```python
scaled_df.head()
```

```python
sns.histplot(df['distance'])
```

```python
sns.histplot(scaled_df['scaled_dist'])
```
